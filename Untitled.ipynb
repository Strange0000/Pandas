```python
data['year']=data['ts'].dt.year
```

```python
this_year=data[data['year']==2024]
last_year=data[data['year']==2023]
```

```python
this_yaer_artist=this_year['artist_name'].value_counts().idxmax()
last_yaer_artist=last_year['artist_name'].value_counts().idxmax()
```

```python
print(f"Most listened artist this year: {this_yaer_artist}")
print(f"Most listened artist last year: {last_yaer_artist}")
```

```python
if this_yaer_artist ==last_yaer_artist:
    print("The most listened artist is the same as last year.")
else:
    print("The most listened artist is different from last year.")
```
