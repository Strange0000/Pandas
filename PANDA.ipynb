```python
mask
```

```python
new_delivery=delivery[mask]
```

## against which 3 team  V kohli score highest run

```python
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
run_scored('V Kohli')
```

```python
def run_scored(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3).index[0]
```

```python
run_scored('MS Dhoni')
```
